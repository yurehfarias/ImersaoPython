```python


df_principal = pd.read_excel("acoes.xlsx", sheet_name="Principal")
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel("acoes.xlsx", sheet_name="Total_de_acoes")
df_total_acoes.head(10)
```

```python
df_tickers = pd.read_excel("acoes.xlsx", sheet_name="Ticker")
df_tickers.head(10)
```

```python
df_chatgpt = pd.read_excel("acoes.xlsx", sheet_name="chatgpt")
df_chatgpt.head(10)
```

```python
df_principal = df_principal[["Ativo",	"Data",	"Último (R$)",	"Var. Dia (%)"]].copy()
df_principal.head(10)
```

```python
df_principal = df_principal.rename(columns= {"Último (R$)":"Valor_final", "Var. Dia (%)":"Var_Dia_pct"}).copy()
df_principal.head(10)
```

```python
df_principal["Var_pct"] = df_principal['Var_Dia_pct'] / 100
df_principal["Valor_inicial"] = df_principal['Valor_final'] / (df_principal["Var_pct"] + 1)
df_principal.head(10)
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on="Ativo", right_on="Código", how="left")
df_principal.head(10)
```

```python
df_principal["Variacao_rs"] = (df_principal['Valor_final'] - df_principal['Valor_inicial']) *  df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = "{:.2f}".format
df_principal
```

```python
df_principal.drop('Código', axis=1, inplace=True)
df_principal
```

```python
df_principal["Qtde. Teórica"] = df_principal["Qtde. Teórica"].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns= {"Qtde. Teórica": "Qtde_teorica"}).copy()
df_principal
```

```python
df_principal["Resultado"] = df_principal["Variacao_rs"].apply(lambda x: 'Subiu' if x > 0 else('Desceu' if x < 0 else 'Estavel'))
df_principal
```

```python
df_principal = df_principal.merge(df_tickers, left_on="Ativo", right_on="Ticker", how="left")
df_principal.drop('Ticker', axis=1, inplace=True)
df_principal.head(10)
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on="Nome", right_on="Nome", how="left")
df_principal.drop('Nome', axis=1, inplace=True)
df_principal.head(10)
```

```python
df_principal.drop(columns=["Idade_x","Ano de Fundação_x","Segmento_x"], inplace=True)
df_principal
```

```python
df_principal = df_principal.rename(columns= {"Idade_y":"Idade"}).copy()
df_principal = df_principal.rename(columns= {"Ano de Fundação_y":"Ano_fundacao"}).copy()
df_principal = df_principal.rename(columns= {"Segmento_y":"Segmento"}).copy()
df_principal
```

```python
df_principal.dropna(subset=["Idade", "Ano_fundacao"], axis=0, inplace=True)
df_principal["Idade"] = df_principal["Idade"].apply(lambda x: int(float(x)))
df_principal["Ano_fundacao"] = df_principal["Ano_fundacao"].apply(lambda x: int(float(x)))
df_principal.head(10)
```

```python
df_principal["Cat_idade"] = df_principal["Idade"].apply(lambda x: 'Mais de 100' if x > 100 else('Menos de 50' if x < 50 else 'Entre 100 e 50'))
df_principal.head(10)
```

```python
maior = df_principal["Variacao_rs"].max()
menor = df_principal["Variacao_rs"].min()
media = df_principal["Variacao_rs"].mean()
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']["Variacao_rs"].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']["Variacao_rs"].mean()

print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Media\tR$ {media:,.2f}")
print(f"Media de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Media de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby("Segmento")["Variacao_rs"].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby("Resultado")["Variacao_rs"].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x="Resultado", y="Variacao_rs", text="Variacao_rs", title="Variação Reais por Resultado")
fig.show()
```
